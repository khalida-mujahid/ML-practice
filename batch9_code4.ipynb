```
def f(bias, weight, x):
  return bias + weight * x
```

```
def sigmoid(x):
  return 1/(1+math.exp(-x))
```

```
w = [0,1]
```

```
r = list()
for x,y in zip(X,Y):
    results = f(w[0], w[1], x)
    final_results = sigmoid(results)
    if final_results > .90:
      r.append(1)
    else:
      r.append(0)
    print(final_results)
    
```

```
from sklearn.datasets import load_boston
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import pandas as pd
import matplotlib.pyplot as plt
```

```
data = load_boston()
df = pd.DataFrame(data.data, columns = data.feature_names)
```

```
df['Label'] = data.target
```

```
import seaborn as sns
```

```
cm = df.corr().round(2)
```

```
plt.figure(figsize=(10,10))
sns.heatmap(data = cm, annot = True)
plt.show()
```

```
plt.scatter(df['RM'], df['Label'])
plt.show()
```

```
model = LinearRegression()
```

```
X = df["RM"]
Y = df['Label']
```

```
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2)
```

```
feature_extractor = PolynomialFeatures(2)
```

```
X_train2 = feature_extractor.fit_transform(X_train.reshape(-1,1))
X_train2.shape
```

```
X_test2 = feature_extractor.transform(X_test.reshape(-1,1))
X_test2.shape
```

```
model.fit(X_train.reshape(-1,1), y_train)
model.score(X_test.reshape(-1,1), y_test)
```

```
model.fit(X_train2, y_train)
model.score(X_test2, y_test)
```
