```
from sklearn.cluster import KMeans
from sklearn.datasets import load_digits
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```
data = load_digits()

X = data.data
y = data.target
```

```
print(X.shape)
```

```
%pdb off
X_reduced_fit = PCA(32).fit_transform(X)
X_reduced = PCA(2).fit_transform(X)
```

```
plt.scatter(X_reduced[:, 0], X_reduced[:, 1], c=y, cmap=plt.cm.Set1)
plt.show()
```

```
model = KMeans(10).fit(X_reduced_fit)
```

```
plt.scatter(X_reduced[:, 0], X_reduced[:, 1], c=model.labels_, cmap=plt.cm.Set1)
plt.show()
```

```
def vis_point(X):
  img = X.reshape(8,8)
  plt.imshow(img, interpolation = 'nearest')
  plt.axis('off')
  plt.show()
```

```
model.cluster_centers_.shape
```

```
for i, centeroid in enumerate(model.cluster_centers_):
  plt.subplot(2,5, i+1)
  img = centeroid.reshape(8,8)
  plt.imshow(img, interpolation = 'nearest')
  plt.axis('off')
  plt.title('Cluster ' + str(i+1))  
plt.show()
```
